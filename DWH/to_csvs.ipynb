```python
def extract_colspecs_from_line(line: str):
    colspecs = []
    in_col = False
    start = 0

    for i, char in enumerate(line + " "):  
        if char != "-" and in_col:
            colspecs.append((start, i+1))
            in_col = False
        elif char == "-" and not in_col:
            start = i
            in_col = True

    return colspecs
```

```python
file_path = "factmatchstats.rpt"
with open(file_path, "r", encoding="utf-8") as f:
    content = f.read()

content = content.replace(",", ".")

with open(file_path, "w", encoding="utf-8") as f:
    f.write(content)

with open("factmatchstats.rpt", "r", encoding="utf-8") as f:
    lines = f.readlines()

colspecs = extract_colspecs_from_line(lines[1])  # druga linia = linia z kreskami
print(colspecs)
df = pd.read_fwf("factmatchstats.rpt", colspecs=colspecs, header=0, skiprows=[1])
df
```

```python
folder_path = "."
rpt_files = [f for f in os.listdir(folder_path) if f.endswith(".rpt")]
dict_files = {}
for rpt_file in rpt_files:
    print(f"Processing file: {rpt_file}")
    file_path = rpt_file
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()
    content = content.replace(",", ".")
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content)
    with open(os.path.join(folder_path, rpt_file), "r", encoding="utf-8") as f:
        lines = f.readlines()
    colspecs = extract_colspecs_from_line(lines[1])  
    df = pd.read_fwf(os.path.join(folder_path, rpt_file), colspecs=colspecs, header=0, skiprows=[1])
    string = rpt_file.split(".")[0]
    #remove two last rows
    df = df.iloc[:-2]
    dict_files[string] = df
    print(df.head())
```

```python
# save all to csv file to dwh_csvs folder
output_folder = "dwh_csvs"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
for key in dict_files:
    output_file = os.path.join(output_folder, f"{key}.csv")
    dict_files[key].to_csv(output_file, index=False)
print(f"Files saved to {output_folder}")
```
